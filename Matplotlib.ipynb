```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([1,2,3],[2,4,6])
plt.title('First Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y, label='2x',color='red',linewidth=2,marker='.',markersize=10,markeredgecolor='green')
plt.title('First Graph 2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([1,2,3])
plt.yticks([2,4,6,8])
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y, label='2x',color='red',linewidth=2,marker='.',markersize=10,markeredgecolor='green',linestyle='--')
plt.title('First Graph 3')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([1,2,3])
plt.yticks([2,4,6,8])
plt.legend()
plt.show()
```

```python
x2=np.arange(0,4,0.5)
print(x2)
```

```python
# Resize the graph
# plt.figure(figsize(5,7),dpi=100)
plt.plot(x2,x2**2,label='long')
plt.plot(x2[:4],x2[:4]**2,color='red',label='short')
plt.legend()
```

```python
labels=['A','B','C']
values=[1,4,2]
bars=plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.show()
```

```python
gas=pd.read_csv('/content/gas_prices.csv')
plt.plot(gas.Year,gas.USA,color='red',label='USA',marker='.',markersize=10)
plt.plot(gas.Year,gas.Canada,color='blue',label='Canada',marker='.',markersize=10)
plt.title('USA v/s Canada Gas Price')
plt.xlabel('Year')
plt.ylabel('Price')
plt.xticks(gas.Year[::3].tolist()+[2011])
plt.legend()
plt.show()
```
